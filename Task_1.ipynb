```python
df_1 = pd.read_excel('Тестовое_задание_для_кандидатов_2022.xlsx', engine='openpyxl', sheet_name = 'Задание 1', skiprows = 4)
df_1.info()
#Удаляем пустые строки, для того чтобы данные были полные
df_1 = df_1.dropna(subset=['Категория устройства'])
df_1.shape
```

```python
df_1['Категория устройства'] = df_1['Категория устройства'].str.lower()
df_1['Сумма начислений в месяц, руб'] = df_1['Сумма начислений в месяц, руб'].round(1)
df_1['Трафик голосовой, мин'] = df_1['Трафик голосовой, мин'].round(1)
df_1['Трафик передачи данных, Мб'] = df_1['Трафик передачи данных, Мб'].round(1)
df_1['Трафик в роуминге, мин'] = df_1['Трафик в роуминге, мин'].round(1)
df_1['Трафик голосовой, дни'] = (df_1['Трафик голосовой, мин']/(60*24)).round(0)
df_1['Трафик передачи данных, Гб'] = (df_1['Трафик передачи данных, Мб']/1024).round(0)
df_1
```

Проверим голосовой трафик на выбросы, будем считать месяц как 30 дней, все что выше уберем из датасета
Принимаем такие условности, так как неуказано может ли с одного ID абонента совершать звонки несколько пользователей
Будем считать что один ID - один пользователь, который физически не может говорить больше 30 дней

```python
df_1 = df_1[df_1['Трафик голосовой, дни'] < 30]
df_1.describe()
```

Удалим трафик передачи данных (Гб) который больше 23 гб (принимаю среднее значение ) и сумма начислений равно 0

```python
df_1 = df_1[~((df_1['Трафик передачи данных, Гб'] > 23) & (df_1['Сумма начислений в месяц, руб'] == 0))]
```

Заменим отрицательные значения в столбцах голосового трафика, трафика передачи данных и трафика в роуминге на медианные значения, так как примем что эти величины не могут быть отрицательные

```python
df_1.loc[df_1["Трафик голосовой, мин"] < 0, "Трафик голосовой, мин"] = \
    df_1.loc[df_1["Трафик голосовой, мин"] >= 0, "Трафик голосовой, мин"].median()
df_1.loc[df_1["Трафик передачи данных, Мб"] < 0, "Трафик передачи данных, Мб"] = \
    df_1.loc[df_1["Трафик передачи данных, Мб"] >= 0, "Трафик передачи данных, Мб"].median()
df_1.loc[df_1["Трафик в роуминге, мин"] < 0, "Трафик в роуминге, мин"] = \
    df_1.loc[df_1["Трафик в роуминге, мин"] >= 0, "Трафик в роуминге, мин"].median()
```
